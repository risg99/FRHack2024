```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
communes = pd.read_excel('files/Communes 41_45_89_pour info.xlsx')
communes.head()
```

```python
measures = pd.read_csv('files/measurements.csv', sep=';')
measures.head()
```

```python
value_counts = measures.Techno.value_counts()

# Plotting
plt.figure(figsize=(6, 5))
bars = value_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Technologies', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Technology Counts', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks() 
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 20, f'{bar.get_height()}', 
             ha='center', va='bottom', fontsize=10)
plt.tight_layout()

# Show plot
plt.show()
```
